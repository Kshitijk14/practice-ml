```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

```python
df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
df
```

```python
df.describe()
```

```python
X = df
y = digits.target
```

```python
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

### Use components such that 95% of variance is retained
-> Retain 95% of useful features.

```python
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

**-> PCA created 29 components out of 64 original columns**

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
model_pca = LogisticRegression(max_iter=1000)
model_pca.fit(X_train_pca, y_train)
```

```python
model_pca.score(X_test_pca, y_test)
```

### Creating PCA by supplying components explicitly

```python
pca = PCA(n_components=2)
X_pca2 = pca.fit_transform(X)
X_pca2
```

```python
X_pca2.shape
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca2, X_test_pca2, y_train, y_test = train_test_split(X_pca2, y, test_size=0.2, random_state=42)

model_pca = LogisticRegression(max_iter=1000)
model_pca.fit(X_train_pca2, y_train)
model_pca.score(X_test_pca2, y_test)
```
