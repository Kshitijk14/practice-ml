## Problem

**Use sklearn.datasets iris flower dataset to train a model using logistic regression. Find accuracy of model and predict different samples in your test dataset. In iris dataset there are 150 samples containing following features:**

1. Sepal Length
2. Sepal Width
3. Petal Length
4. Petal Width

**Using above 4 features you will classify a flower in one of the three categories:**

1. Setosa
2. Versicolour
3. Virginica

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
iris
```

```python
dir(iris)
```

```python
iris.keys()
```

```python
iris.feature_names
```

```python
list(iris.target_names)
```

```python
iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
model.score(X_test, y_test)
```
