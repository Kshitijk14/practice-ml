**Problem Statement:** Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
```

```python
new_df = df.drop('per capita income (US$)', axis='columns')
new_df.head(5)
```

```python
income = df['per capita income (US$)']
income.head(5)
```

```python
reg = linear_model.LinearRegression()
reg
```

```python
reg.fit(new_df, income)
```

(1) Predict the per capita income in the year = 2020

```python
reg.predict([[2020]])
```

(1) The per capita income in the year 2020 = 41288.69409442 (US$)

```python
y_pred = reg.predict(new_df)
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
plt.plot(df['year'], y_pred, color='blue')
```
