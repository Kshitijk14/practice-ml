```python
import pandas as pd

df = pd.read_csv('bhp.csv')
df
```

```python
df.describe()
```

Explore samples that are above 99.90% percentile and below 1% percentile rank

```python
min_threshold, max_threshold = df.price_per_sqft.quantile([0.01, 0.999])
min_threshold, max_threshold
```

```python
df[df.price_per_sqft < min_threshold]
```

```python
df[df.price_per_sqft < min_threshold].shape
```

```python
df[df.price_per_sqft > max_threshold]
```

```python
df[df.price_per_sqft > max_threshold].shape
```

## Remove outliers

```python
df = df[(df.price_per_sqft<max_threshold) & (df.price_per_sqft>min_threshold)]
df.shape
```
