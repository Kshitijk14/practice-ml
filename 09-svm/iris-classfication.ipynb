## Data Analysis & Preprocessing

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
# Add the target as a new column
df['target'] = iris.target
df
```

```python
df.select_dtypes(include=[np.number]).groupby('target').mean()
```

```python
# Map the target values to target names

# df['flower_names'] = df.target.map({i: name for i, name in enumerate(iris['target_names'])})

df['flower_names'] =df.target.apply(lambda x: iris.target_names[x])
df
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

## Model

```python
X = df.drop(['target', 'flower_names'], axis='columns')
X
```

```python
y = df['target']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
X_train
```

```python
X_test
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

y_pred = model.predict(X_test)
cf = confusion_matrix(y_test, y_pred)

sns.heatmap(cf, annot=True, fmt="")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```

## Tune parameters

**1. Regularization (C):** [high value C i.e. high regularization]

```python
# C = 1 (default)
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
# C -> hit n trial
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

**2. Gamma:**

```python
# gamma = auto deprecated (default)
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

**3. Kernel:**

```python
# kernel = rbf (default)
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
